```python
# Data manipulation
# ==============================================================================
import pandas as pd
import numpy as np
import datetime

# Plots
# ==============================================================================
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Modelling and Forecasting
# ==============================================================================
import pyflux as pf

# Warnings configuration
# ==============================================================================
import warnings
warnings.filterwarnings("ignore")

# Yahoo finance is used to fetch data 
# ==============================================================================
import yfinance as yf
yf.pdr_override()
```

```python
symbol = '^GSPC' # input
start = '2020-01-01' # input 
end = '2022-07-18'  # input
df = yf.download(symbol, start, end)
```

```python
df.head()
```

```python
df.tail()
```

```python
log_returns = pd.DataFrame(np.diff(np.log(df['Adj Close'].values)))
log_returns.index = df.index.values[1:df.index.values.shape[0]]
log_returns.columns = [symbol + " Returns"]

log_returns.head()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(log_returns)
plt.ylabel("Returns")
plt.title(symbol + " Returns")
plt.show()
```

```python
pf.acf_plot(log_returns.values.T[0])
pf.acf_plot(np.square(log_returns.values.T[0]))
```

```python
my_model = pf.GARCH(p=1, q=1, data=log_returns)
print(my_model.latent_variables)
```

```python
result = my_model.fit('M-H', nsims=20000)
```

```python
my_model.plot_z([1,2])
```

```python
my_model.plot_fit(figsize=(15,5))
```

```python
my_model.plot_sample(nsims=10, figsize=(15,7))
```

```python
from scipy.stats import kurtosis
my_model.plot_ppc(T=kurtosis)
```

```python
my_model.plot_predict(h=30, figsize=(15,5))
```
