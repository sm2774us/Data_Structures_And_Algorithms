```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt

# yahoo finance is used to fetch data 
import yfinance as yf
yf.pdr_override()
```

```python
# input
symbol = 'AAPL'
start = '2018-01-01'
end = '2019-01-01'

# Read data 
df = yf.download(symbol,start,end)

# View Columns
df.head()
```

```python
df.tail()
```

```python
df['returns'] = df['Adj Close'].pct_change()
df = df.dropna()
```

```python
# Extract the closing prices from the data
df['sin'] = np.sin(df['Adj Close'])
df['cos'] = np.cos(df['Adj Close'])
df['tan'] = np.tan(df['Adj Close'])
```

```python
df.head()
```

```python
df.tail()
```

```python
df[['returns', 'sin']].plot(figsize=(10,6))
plt.title(symbol + ' sin')

# Show the plot
plt.show()
```

```python
df[['returns', 'cos']].plot(figsize=(10,6))
plt.title(symbol + ' cos')

# Show the plot
plt.show()
```

```python
df[['returns', 'tan']].plot(figsize=(10,6))
plt.title(symbol + ' tan')

# Show the plot
plt.show()
```

```python
# Extract the returns prices from the data
df['sin_r'] = np.sin(df['returns'])
df['cos_r'] = np.cos(df['returns'])
df['tan_r'] = np.tan(df['returns'])
```

```python
df[['returns', 'sin_r']].plot(figsize=(10,6))
plt.title(symbol + ' sin returns')

# Show the plot
plt.show()
```

```python
df[['returns', 'cos_r']].plot(figsize=(10,6))
plt.title(symbol + ' cos returns')

# Show the plot
plt.show()
```

```python
df[['returns', 'tan_r']].plot(figsize=(10,6))
plt.title(symbol + ' tan returns')

# Show the plot
plt.show()
```

```python
# Extract 'Adj Close' column
adj_close = df['Adj Close']

# Apply trigonometric functions
sin_values = np.sin(adj_close)
cos_values = np.cos(adj_close)
tan_values = np.tan(adj_close)
arctan_values = np.arctan(adj_close)

# Plotting the results
plt.figure(figsize=(16, 8))
plt.subplot(2, 2, 1)
plt.plot(adj_close)
plt.title('Adj Close')

plt.subplot(2, 2, 2)
plt.plot(sin_values)
plt.title('Sin(Adj Close)')

plt.subplot(2, 2, 3)
plt.plot(cos_values)
plt.title('Cos(Adj Close)')

plt.subplot(2, 2, 4)
plt.plot(arctan_values)
plt.title('Arctan(Adj Close)')

plt.tight_layout()
plt.show()
```
