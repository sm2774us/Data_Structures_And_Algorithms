# Poisson distribution

https://en.wikipedia.org/wiki/Poisson_distribution

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close']

# Create a Poisson distribution
lambda_value = 3  # Adjust this parameter to control the distribution
poisson_dist = poisson(mu=lambda_value)

# Generate random samples from the Poisson distribution
sample_size = len(data)
poisson_samples = poisson_dist.rvs(size=sample_size)

# Add Poisson samples to the stock data
data_with_poisson = data + poisson_samples

# Plot the original and modified data
plt.figure(figsize=(12, 6))
plt.plot(data, label='Original Stock Data', linewidth=2)
plt.plot(data_with_poisson, label='Stock Data with Poisson Noise', linewidth=2)
plt.legend()
plt.title('Stock Data with Poisson Noise')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Plot the original data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(data, label='Original Stock Data', linewidth=2)
plt.legend()
plt.title('Original Stock Data')
plt.xlabel('Date')
plt.ylabel('Price')

# Plot the modified data with Poisson noise
plt.subplot(1, 2, 2)
plt.plot(data_with_poisson, label='Stock Data with Poisson Noise', linewidth=2)
plt.legend()
plt.title('Stock Data with Poisson Noise')
plt.xlabel('Date')
plt.ylabel('Price')

plt.tight_layout()
plt.show()
```
