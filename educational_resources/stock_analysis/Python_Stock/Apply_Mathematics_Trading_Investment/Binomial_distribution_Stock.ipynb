# Binomial distribution

https://en.wikipedia.org/wiki/Binomial_distribution

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a function for the binomial distribution
def binomial_distribution(prob_success, num_trials, size):
    """Simulate a binomial distribution.

    Parameters:
    prob_success (float): Probability of success for each trial.
    num_trials (int): Number of trials.
    size (int): Number of random samples to generate.

    Returns:
    ndarray: An array of random samples following a binomial distribution.
    """
    return np.random.binomial(num_trials, prob_success, size)

# input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Read data 
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close']

# Define parameters for the binomial distribution
prob_success = 0.5  # Probability of success for each trial (you can adjust this)
num_trials = 10    # Number of trials (you can adjust this)
size = len(data)    # Number of samples equal to the length of your stock data

# Simulate the binomial distribution
binomial_samples = binomial_distribution(prob_success, num_trials, size)

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(data.index, binomial_samples, label='Binomial Distribution', color='blue')
plt.title('Simulated Binomial Distribution with Stock Data')
plt.xlabel('Date')
plt.ylabel('Binomial Samples')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the stock data in one chart
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(data.index, data, label='Stock Price')
plt.title(f'{symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Plot the Bernoulli trials in a separate chart
plt.subplot(2, 1, 2)
plt.plot(data.index, binomial_samples, label='Binomial Distribution', color='red')
plt.title('Bernoulli Trials')
plt.xlabel('Date')
plt.ylabel('Outcome')
plt.legend()

plt.tight_layout()  # Ensures proper spacing between subplots

plt.show()
```

```python
# Convert the binomial_samples array into a DataFrame
binomial_df = pd.DataFrame(binomial_samples, index=data.index, columns=['Binomial Samples'])

# Print the DataFrame
print(binomial_df)
```
