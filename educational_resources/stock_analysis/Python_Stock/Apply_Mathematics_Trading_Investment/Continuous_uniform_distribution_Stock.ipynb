# Continuous uniform distribution

https://en.wikipedia.org/wiki/Continuous_uniform_distribution

```python
import warnings
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close'].values

# Define the lower and upper bounds for the uniform distribution
a = min(data)  # Minimum value in the data
b = max(data)  # Maximum value in the data

# Calculate the CDF for the data
def uniform_cdf(x):
    if x < a:
        return 0.0
    elif x >= b:
        return 1.0
    else:
        return (x - a) / (b - a)

# Create a range of x values for the CDF
x_values = np.linspace(a, b, 100)

# Calculate the CDF values
cdf_values = [uniform_cdf(x) for x in x_values]

# Plot the CDF
plt.plot(x_values, cdf_values, label='CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Financial Data')
plt.legend()
plt.grid()
plt.show()
```

```python
import warnings
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close'].values

# Calculate the minimum and maximum values of the data
a = np.min(data)
b = np.max(data)

# Calculate the CDF
def continuous_uniform_cdf(x, a, b):
    if x < a:
        return 0
    elif x > b:
        return 1
    else:
        return (x - a) / (b - a)

# Create a range of values for x
x_values = np.linspace(a, b, 100)

# Calculate the CDF values for the range of x
cdf_values = [continuous_uniform_cdf(x, a, b) for x in x_values]

# Plot the CDF
plt.plot(x_values, cdf_values, label='CDF')
plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate the range for the uniform distribution
min_value = np.min(data)
max_value = np.max(data)

# Generate random data following a continuous uniform distribution within the same range
random_uniform_data = np.random.uniform(min_value, max_value, size=len(data))

# Create a histogram for the actual data and the generated data
plt.figure(figsize=(10, 5))
plt.hist(data, bins=50, alpha=0.5, label='Actual Data', color='blue')
plt.hist(random_uniform_data, bins=50, alpha=0.5, label='Uniform Distribution', color='green')
plt.xlabel('Adj Close Prices')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram of "Adj Close" Prices with Continuous Uniform Distribution')
plt.show()
```
