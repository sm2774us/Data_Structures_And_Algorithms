# One-way wave equation

https://en.wikipedia.org/wiki/One-way_wave_equation

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define the one-way wave equation function
def one_way_wave_equation(x, A, k, omega, phi):
    return A * np.sin(k * x - omega) + phi

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close'].values

# Create x values for the wave (assuming uniform time intervals)
x = np.arange(len(data))

# Parameters for the wave
A = np.max(data)  # Amplitude
k = 2 * np.pi / len(data)  # Wave number
omega = 0  # Phase offset
phi = np.mean(data)  # Vertical shift

# Calculate the wave
wave = one_way_wave_equation(x, A, k, omega, phi)

# Plot the stock data and the wave
plt.figure(figsize=(12, 6))
plt.plot(x, data, label='Stock Data', color='b')
plt.plot(x, wave, label='Wave', color='r')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.title(f'Stock Data and One-Way Wave for {symbol}')
plt.grid(True)
plt.show()
```
