# Discrete uniform distribution

https://en.wikipedia.org/wiki/Discrete_uniform_distribution  

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define a function for generating a discrete uniform distribution
def generate_discrete_uniform(low, high, size):
    return np.random.randint(low, high + 1, size)

# Input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close']

# Define the range for the discrete uniform distribution (e.g., 0 to 100)
low = 0
high = 100

# Generate random values from the discrete uniform distribution
num_samples = len(data)
uniform_samples = generate_discrete_uniform(low, high, num_samples)

# Apply the generated distribution to the stock data
data_with_uniform_noise = data + uniform_samples

# Plot the original and modified data
plt.figure(figsize=(12, 6))
plt.plot(data.index, data, label='Original Data', linewidth=2)
plt.plot(data.index, data_with_uniform_noise, label='Data with Uniform Noise', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title('Stock Data with Discrete Uniform Noise')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the original and modified data as bar charts
plt.figure(figsize=(12, 6))
plt.bar(data.index, data, label='Original Data', width=0.4, align='center', alpha=0.6, color='b')
plt.bar(data.index, data_with_uniform_noise, label='Data with Uniform Noise', width=0.4, align='edge', alpha=0.6, color='r')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title('Stock Data with Discrete Uniform Noise')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate a discrete uniform distribution
def discrete_uniform(low, high, size=1):
    return np.random.randint(low, high + 1, size)

# Input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close']

# Define the parameters for the discrete uniform distribution
low = dataset['Adj Close'].min()  # The lowest value
high = dataset['Adj Close'].max()  # The highest value

# Generate random numbers from the discrete uniform distribution
random_numbers = discrete_uniform(low, high, size=len(data))

# Plot the random numbers against the stock data
plt.figure(figsize=(12, 6))
plt.plot(data.index, data, label='Stock Price')
plt.plot(data.index, random_numbers, label='Random Numbers', alpha=0.7)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Stock Price vs. Random Numbers (Discrete Uniform Distribution)')
plt.legend()
plt.show()
```

```python
# Create the first chart (Stock Price)
plt.figure(figsize=(12, 8))

# Subplot for Stock Price
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot
plt.plot(data.index, data, label='Stock Price')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Stock Price')
plt.legend()

# Subplot for Random Numbers
plt.subplot(2, 1, 2)  # 2 rows, 1 column, second subplot
plt.plot(data.index, random_numbers, label='Random Numbers', alpha=0.7)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Random Numbers (Discrete Uniform Distribution)')
plt.legend()

plt.tight_layout()  # Ensures that the subplots do not overlap
plt.show()
```
