# de Villiers (1977, 1980) model equation

https://www.ncbi.nlm.nih.gov/pmc/articles/PMC1332909/pdf/jeabehav00084-0019.pdf

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the generic de Villiers equation (replace with the specific equation)
def de_villiers_equation(x, a, b, c):
    return a * x**2 + b * x + c

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
data = dataset['Adj Close'].values
x = np.arange(len(data))  # Create an array of x values (time)

# Fit the equation to the data
popt, _ = curve_fit(de_villiers_equation, x, data)

# Extract the optimized parameters
a, b, c = popt

# Predict the values using the fitted equation
predicted_data = de_villiers_equation(x, a, b, c)

# Plot the original data and the fitted curve
plt.figure(figsize=(10, 5))
plt.plot(x, data, 'b.', label='Original Data')
plt.plot(x, predicted_data, 'r-', label='Fitted Curve')
plt.xlabel('Time')
plt.ylabel('Adj Close Price')
plt.legend()
plt.show()

# Print the optimized parameters
print("Optimized Parameters (a, b, c):", a, b, c)
```
