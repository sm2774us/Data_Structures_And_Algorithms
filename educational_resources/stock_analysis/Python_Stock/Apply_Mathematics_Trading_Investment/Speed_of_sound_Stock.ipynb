# Speed of Sound

https://en.wikipedia.org/wiki/Speed_of_sound

```python
import warnings
warnings.filterwarnings("ignore")
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Speed of sound function
def speed_of_sound(pressure, density):
    gamma = 1.4  # Adiabatic index for air
    speed = np.sqrt(gamma * pressure / density)
    return speed

# Input
symbol = 'AMD'
start = '2020-01-01'
end = '2023-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Prepare data
stock_prices = dataset['Adj Close'].values

# Use a constant value for density (you can change this value as needed)
density = 1.225  # Approximate density of air at sea level in kg/m^3

# Calculate the speed of sound for the stock data
speeds = [speed_of_sound(price, density) for price in stock_prices]

# Create a DataFrame to store the results
speed_data = pd.DataFrame({'Date': dataset.index, 'Speed of Sound': speeds})

# Plot the speed of sound
plt.figure(figsize=(12, 6))
plt.plot(speed_data['Date'], speed_data['Speed of Sound'], label='Speed of Sound', color='blue')
plt.title(f'Speed of Sound for {symbol} Stock (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Speed of Sound')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the speed of sound and Original Price
plt.figure(figsize=(12, 6))
plt.plot(dataset['Adj Close'], color='blue', label='Price')
plt.plot(speed_data['Date'], speed_data['Speed of Sound'], label='Speed of Sound', color='red')
plt.title(f'Speed of Sound for {symbol} Stock (2020-2023)')
plt.xlabel('Date')
plt.ylabel('Speed of Sound')
plt.legend()
plt.grid(True)
plt.show()
```
