# Law of the Iterated Logarithm

https://en.wikipedia.org/wiki/Law_of_the_iterated_logarithm

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define a function to calculate Law of the Iterated Logarithm bounds
def law_of_iterated_logarithm(data):
    n = len(data)
    mean = np.mean(data)
    log_n = np.log(np.log(n))
    epsilon = np.sqrt(2 * log_n / n)

    upper_bound = mean + epsilon
    lower_bound = mean - epsilon

    return upper_bound, lower_bound

# Input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Read data
dataset = yf.download(symbol, start, end)

# Extract the 'Close' prices
closing_prices = dataset['Close']

# Calculate Law of the Iterated Logarithm bounds
upper_bound, lower_bound = law_of_iterated_logarithm(closing_prices)

# Plot the stock prices and the LIL bounds
plt.figure(figsize=(12, 6))
plt.plot(closing_prices, label='Stock Price')
plt.axhline(upper_bound, color='r', linestyle='--', label='Upper Bound (LIL)')
plt.axhline(lower_bound, color='g', linestyle='--', label='Lower Bound (LIL)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{symbol} Stock Price with Law of the Iterated Logarithm Bounds')
plt.legend()
plt.grid(True)
plt.show()
```
