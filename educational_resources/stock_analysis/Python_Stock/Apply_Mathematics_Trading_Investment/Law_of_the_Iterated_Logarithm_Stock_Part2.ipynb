# Law of the Iterated Logarithm

https://en.wikipedia.org/wiki/Law_of_the_iterated_logarithm

```python
import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def law_of_iterated_logarithm(symbol, start, end):
    # Download stock data
    dataset = yf.download(symbol, start, end)
    
    # Calculate daily returns
    dataset['Daily Returns'] = dataset['Adj Close'].pct_change()
    
    # Calculate cumulative sum of daily returns
    # dataset['Cumulative Returns'] = (1 + dataset['Daily Returns']).cumprod() - 1
    
    # Calculate upper and lower bounds based on the law of the iterated logarithm
    n = len(dataset)
    upper_bound = np.sqrt(2 * np.log(np.log(n))) / np.sqrt(n)
    lower_bound = -np.sqrt(2 * np.log(np.log(n))) / np.sqrt(n)
    
    # Plot the daily returns and bounds
    plt.figure(figsize=(12, 6))
    plt.plot(dataset.index, dataset['Daily Returns'], label='Daily Returns', color='blue')
    plt.axhline(upper_bound, color='red', linestyle='--', label='Upper Bound')
    plt.axhline(lower_bound, color='green', linestyle='--', label='Lower Bound')
    plt.xlabel('Date')
    plt.ylabel('Daily Returns')
    plt.title(f'{symbol} Stock Price - Law of the Iterated Logarithm')
    plt.legend()
    plt.grid(True)
    plt.show()

# Input
symbol = 'AMD'
start = '2014-01-01'
end = '2019-01-01'

# Apply the law of the iterated logarithm to the stock data
law_of_iterated_logarithm(symbol, start, end)
```
