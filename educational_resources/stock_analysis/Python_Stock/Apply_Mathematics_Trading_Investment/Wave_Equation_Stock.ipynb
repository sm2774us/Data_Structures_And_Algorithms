```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt

# yahoo finance is used to fetch data 
import yfinance as yf
yf.pdr_override()
```

```python
# input
symbol = 'AAPL'
start = '2018-01-01'
end = '2019-01-01'

# Read data 
df = yf.download(symbol,start,end)

# View Columns
df.head()
```

```python
df.tail()
```

```python
# Extract the closing prices from the data
prices = df['Adj Close']
```

```python
# Calculate the wave equation values
time = np.arange(len(prices))
amplitude = (prices - np.mean(prices)) / np.std(prices)
wave = amplitude * np.sin(2 * np.pi * time / len(prices))
```

```python
# Plot figure and axes
fig, ax = plt.subplots()

# Plot the wave equation
ax.plot(time, wave)

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Wave')
ax.set_title(symbol + ' Wave Equation')

# Show the plot
plt.show()
```

```python
# Normalize the data
normalize = (prices - prices.min())/ (prices.max() - prices.min())
```

```python
# Plot figure and axes
fig, ax = plt.subplots(figsize=(12,8))

# Plot the wave equation
ax.plot(time, wave, label='Wave')
ax.plot(time, normalize, label='Normalize Price')

# Set labels and title
ax.legend()
ax.set_xlabel('Time')
ax.set_ylabel('Wave')
ax.set_title(symbol + ' Wave Equation Comparison')

# Show the plot
plt.show()
```
