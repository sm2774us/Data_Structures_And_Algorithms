# Bears Power

https://www.motivewave.com/studies/bears_power.htm 

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# yahoo finance is used to fetch data 
import yfinance as yf
yf.pdr_override()
```

```python
# input
symbol = 'AAPL'
start = '2019-01-01'
end = '2020-01-01'

# Read data 
df = yf.download(symbol,start,end)

# View Columns
df.head()
```

```python
n = 14 # Number of Periods
df['BearsPower'] = df['Low'] - pd.Series(df['Adj Close'].rolling(n).mean())
```

```python
df = df.dropna()
df.head()
```

```python
df.tail()
```

```python
fig = plt.figure(figsize=(14,10))
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df['Adj Close'])
ax1.set_title('Stock '+ symbol +' Closing Price')
ax1.set_ylabel('Price')

ax2 = plt.subplot(2, 1, 2)
ax2.plot(df['BearsPower'], label='Bears Power', color='red')
ax2.axhline(y=df['BearsPower'].mean(), color='blue', linestyle='--')
ax2.axhline(y=df['BearsPower'].min(), color='green', linestyle='--')
ax2.axhline(y=df['BearsPower'].max(), color='red', linestyle='--')
ax2.grid()
ax2.set_ylabel('Bears Power')
ax2.set_xlabel('Date')
ax2.legend(loc='best')
```

```python
fig = plt.figure(figsize=(14,10))
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df['Adj Close'])
ax1.set_title('Stock '+ symbol +' Closing Price')
ax1.set_ylabel('Price')

ax2 = plt.subplot(2, 1, 2)
ax2.plot(df['BearsPower'], label='Bears Power', color='k')
ax2.fill_between(df.index, y1=df['BearsPower'], y2=0, where=df['BearsPower']>0, color='green')
ax2.fill_between(df.index, y1=df['BearsPower'], y2=0, where=df['BearsPower']<0, color='red')
ax2.grid()
ax2.set_ylabel('Bears Power')
ax2.set_xlabel('Date')
```
