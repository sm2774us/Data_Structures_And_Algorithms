# Arms Ease of Movement

https://www.linnsoft.com/techind/arms-ease-movement

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# yahoo finance is used to fetch data 
import yfinance as yf
yf.pdr_override()
```

```python
# input
symbol = 'AAPL'
start = '2018-01-01'
end = '2019-01-01'

# Read data 
df = yf.download(symbol,start,end)
df.index = pd.Index(df.index.date, name="Date")

# View Columns
df.head()
```

```python
def emv(df, n):
    distanceMoved = ((df['High'] + df['Low']) / 2) - ((df['High'].shift(1) + df['Low'].shift(1)) / 2)
    boxRatio = (df['Volume'] / 100000000) / ((df['High'] - df['Low']) / 1000)
    emv = distanceMoved / boxRatio
    emvMA = emv.rolling(n).mean()
    return emvMA
```

```python

n = 20 # choose the number of periods for the moving average
```

```python
df['EMV'] = emv(df, n)
df['ArmsEMV'] = df['EMV'] / df['Adj Close']
```

```python
df = df.dropna()
df.head()
```

```python
fig = plt.figure(figsize=(16,10))
ax1 = plt.subplot(2, 1, 1)
ax1.plot(df['Adj Close'])
ax1.set_title('Stock '+ symbol +' Closing Price')
ax1.set_ylabel('Price')
ax1.tick_params(axis='x', rotation=45)

ax2 = plt.subplot(2, 1, 2)
ax2.plot(df['ArmsEMV'] , label='Arms Ease of Movement', color='red')
ax2.axhline(y=df['ArmsEMV'].mean(), color='blue', linestyle='--')
ax2.grid()
ax2.set_ylabel('Arms Ease of Movement')
ax2.set_xlabel('Date')
ax2.tick_params(axis='x', rotation=45)
ax2.legend(loc='best')
```
